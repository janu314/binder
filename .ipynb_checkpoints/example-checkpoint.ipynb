# Binder example
import pandas as pd
import matplotlib.pyplot as plt


```python
# Generate sample data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [100, 150, 200, 250, 300]
}

df = pd.DataFrame(data)
df.head()
```

```python
# Plotting
plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Plot Over Time')
plt.grid(True)
plt.show()
```
